```python
import pandas as pd
#Read data
df = pd.read_csv("casestudy.csv")
df.head()
```

```python
df.info()
```

```python
import numpy as np

# total revenue for year = 2015
year =2015

cur_df = df.loc[df.year==year]
prev_df = df.loc[df.year==year]

total_revenue = np.sum(cur_df.net_revenue)
print('Total revenue for', year,': ',total_revenue)

# new_customer revenue
df_new_customer = cur_df.loc[cur_df.customer_email.isin(prev_df.customer_email)==False]
new_customer_revenue = np.sum(df_new_customer.net_revenue)
print('New customer revenue for', year,': ',new_customer_revenue)


#Existing Customer Revenue Current Year
df_exist_customer = cur_df.loc[cur_df.customer_email.isin(prev_df.customer_email)]
existing_customer_revenue_cur = np.sum(df_exist_customer.net_revenue)
print('Existing Customer Revenue Current Year', year,': ',existing_customer_revenue_cur)

#Existing Customer Revenue Prior Year
df_exist_customer_prior = prev_df.loc[prev_df.customer_email.isin(cur_df.customer_email)]
existing_customer_revenue_prior = np.sum(df_exist_customer_prior.net_revenue)
print('Existing Customer Revenue Prior Year', year-1,': ',existing_customer_revenue_prior)

# Existing customer growth, Revenue of existing customers for current year – Revenue of existing customers from existing year
existing_customer_growth = existing_customer_revenue_cur - existing_customer_revenue_prior
print('Existing customer growth for', year,': ',customer_growth)

#Revenue lost from attrition
df_lost_customers = prev_df.loc[prev_df.customer_email.isin(cur_df.customer_email)==False]
revenue_lost =  np.sum(df_lost_customers.net_revenue)
print('Revenue lost from attrition', year,': ',revenue_lost)
#Total Customers Current Year
total_cur_customers=cur_df.size
print('Total Customers Current Year', year,': ',total_cur_customers)

#Total Customers Previous Year
total_prev_customers=prev_df.size
print('Total Customers Previous Year', year-1,': ',total_prev_customers)

#New Customers
new_customers = df_new_customer.size
print('New Customers number for', year,': ',new_customers)
#Lost Customers
lost_customers = df_lost_customers.size
print('Lost Customers number for', year,': ',lost_customers)
```
